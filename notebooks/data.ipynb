```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['svg.fonttype'] = 'none'
```

```python
def plot_loss(df: pd.DataFrame):
    labels = ["steer", "seq2seq", "pilotnet"]

    # Clip the values in the DataFrame
    for label in labels:
        df[label] = df[label].clip(upper=1000)

    plt.figure(figsize=(10,5))
    plt.xlabel("step")
    plt.ylabel("loss value")

    for label in labels:
        plt.plot(df["step"], df[label], label=label, linewidth=2)

    plt.legend()
```

```python
train_loss_df = pd.read_csv("./graphs/loss_train.csv")

plot_loss(train_loss_df)
```

```python
val_loss_df = pd.read_csv("./graphs/loss_val.csv")

plot_loss(val_loss_df)
plt.savefig("val_loss.pdf", bbox_inches='tight')
```
